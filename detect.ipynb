```python
import requests
import json
request_pokemon = requests.get("https://pokeapi.co/api/v2/pokemon?limit=151=0")
request_pokemon = request_pokemon.json()
```

```python
print(request_pokemon)
```

```python
parsed_data = request_pokemon


for item in parsed_data['results']:
    url = item['url']

    response = requests.get(url)
    
    if response.status_code == 200:
        pokemon_data = response.json()
        pokemon_name = item['name']
        file_name = f"json/gen1/{pokemon_name}.json"
        
        with open(file_name, 'w') as file:
            json.dump(pokemon_data, file)
            print(f"Đã lưu dữ liệu từ {url} vào tệp {file_name}")
    else:
        print(f"Không thể lấy dữ liệu từ {url}")
```

```python
import csv
import json
for json_nfile in range(10,11):
    with open(f'json/pokedex_{json_nfile}.json','r') as json_file:
        pokemon_data = json.load(json_file)
        
    csv_filename = 'csv/pokemon.csv'
    # Define the fieldnames for the CSV file
    fieldnames = ["id", "name", "height", "weight", "xp"] + [stat["name"] for stat in pokemon_data[0]["stats"]]

    # Open the CSV file in write mode
    with open(csv_filename, mode='a', newline='', encoding='utf-8') as csv_file:
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        # Write the header
        writer.writeheader()

        # Write each Pokemon's data to the CSV file
        for pokemon in pokemon_data:
            stats = {stat['name']: stat['base_stat'] for stat in pokemon['stats']}

            writer.writerow({
                "id": pokemon["id"],
                "name": pokemon["name"],
                "height": pokemon["height"],
                "weight": pokemon["weight"],
                "xp": pokemon["xp"],
                **stats
            })

    print(f"Data has been written to {csv_filename} {json_nfile}")
```
